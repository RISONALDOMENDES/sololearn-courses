```python
import yfinance as yf
```

```python
data = yf.Ticker('ITSA4.SA')
price = data.history(period= '4y')
```

```python
x = price['Close'].pct_change()
print(x)
```

```python
x.plot(kind='hist')
```

```python
returns = (x + 1).cumprod()
returns.plot()
```
