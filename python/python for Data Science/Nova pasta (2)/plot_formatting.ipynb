```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('.\\ca-covid.csv')

df.drop('state', axis=1, inplace=True)

df['date'] = pd.to_datetime(df['date'], format= "%d.%m.%y")
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day

df.set_index('day', inplace=True)

df = df[df['month'] == 6]

df[['cases', 'deaths']].plot(kind="line", legend=True)

plt.xlabel('Days in june')
plt.ylabel('Number')
plt.savefig('plot.png')
```

```python
# colocando cor no grafico

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('.\\ca-covid.csv')

df.drop('state', axis=1, inplace=True)

df['date'] = pd.to_datetime(df['date'], format= "%d.%m.%y")
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day

df.set_index('day', inplace=True)

df = df[df['month'] == 6]
# kind mostra o tipo de grafico,
df[['cases', 'deaths']].plot(kind="area", legend=True, stacked=False, color=['#1970E7', '#E73E19'])

plt.xlabel('Days in june')
plt.ylabel('Number')
plt.savefig('plot.png')
```
