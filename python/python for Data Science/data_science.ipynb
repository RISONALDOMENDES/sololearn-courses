MATH OPERATIONS WITH NUMPY

houses prices

```python
# House Prices

import numpy as np

data = np.array([150000, 125000, 320000, 540000, 200000, 120000, 160000, 230000, 280000, 290000, 300000, 500000, 420000, 100000, 150000, 280000])

media = np.mean(data)
desvio = np.std(data)

dmin = media - desvio
dmax = media + desvio

quant = 0
for x in data:
    if (x > dmin) & (x < dmax):
        quant += 1
        
print((quant / len(data))* 100)
```

DATA MANIPULATION WITH PANDAS

covid data analysis

```python
# covid data analysis

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ca-covid.csv')

df.head()
```

```python
df.drop('state', axis=1, inplace=True)
df.set_index('date', inplace=True)

df['ratio'] = df['deaths'] / df['cases']

print(df[df['ratio'] == df['ratio'].max()])
```

```python
import matplotlib.pyplot as plt

df.plot(kind='bar')
plt.savefig('plt.png')
```

```python
df.plot(kind='line')
```

```python
df.plot(kind='box')
```

```python
df.plot(kind='hist')
```

```python
df[['cases', 'deaths']].plot(kind="line",legend= True)

plt.xlabel('Days in June')
plt.ylabel('Number')
```

```python
plt.suptitle("COVID-19 IN JUNE")
df[['cases', 'deaths']].plot(kind="area", legend=True, stacked=False, color=['#1970E7', '#E73E19'])
```

```python
df.drop('state', axis=1, inplace=True)
df['date'] = pd.to_datetime(df['date'], format="%d.%m.%y")
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df.set_index('day', inplace=True)

df = df[df['month']==6]
        

df[['cases', 'deaths']].plot(kind="area", legend=True, stacked=False, color=['#1970E7', '#E73E19'])

plt.xlabel('Days in June')
plt.ylabel('Number')
plt.suptitle("COVID-19 IN JUNE")
plt.savefig('plot.png')
```
